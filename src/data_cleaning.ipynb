```python
import pandas as pd
```

```python
df = pd.read_csv('..\data\FAA Wildlife Strikes.csv')
```

```python
df.iloc[ : , :7]
```

```python
df['Aircraft: Number of engines'] = pd.to_numeric(df['Aircraft: Number of engines'], errors='coerce').astype(pd.Int64Dtype())
```

```python
df.iloc[ : , 6:13]
```

```python
df['Collision Date and Time'] = pd.to_datetime(df['Collision Date and Time'])
```

```python
df['Collision Date and Time'] = pd.to_datetime(df['Collision Date and Time'])
df['Day of Year'] = df['Collision Date and Time'].dt.dayofyear
```

```python
df.iloc[ : , 12:19]
```

```python
feet_to_meters = 0.3048
df['m above ground'] = df['Feet above ground'] * feet_to_meters
```

```python
miles_to_km = 1.60934
df['km from airport'] = df['Miles from airport'] * miles_to_km
```

```python
df['Cost: Total $'] = df['Cost: Total $'].str.replace(',', '')
```

```python
df['Cost: Total $'] = df['Cost: Total $'].astype(int)
```

```python
df['Cost: Total $'] = pd.to_numeric(df['Cost: Total $'], errors='coerce').astype(pd.Int64Dtype())
```

```python
df['Feet above ground'] = df['Feet above ground'].str.replace(',', '')
df['Feet above ground'] = pd.to_numeric(df['Feet above ground'], errors='coerce')
```

```python
df.iloc[ : , 18:]
```

```python
null_records = df[df['Record ID'].isnull()]
non_unique_records = df[df.duplicated(subset=['Record ID'], keep=False)]

print("NULL Records:")
print(null_records)

print("\nNon-Unique Records:")
print(non_unique_records)
```

```python
df.set_index('Record ID', inplace=True)  
df.to_csv('../data/FAA_wildlife_strikes_cleaned.csv', index=True)
```
